````python
# ================================
# PROYECTO AQUA LIMPIA S.A.
# Análisis exploratorio de plantas
# ================================

import pandas as pd
import plotly.express as px
from google.colab import files

# Cargar archivo
uploaded = files.upload()

# Leer dataset
df = pd.read_excel("dataset_set_A_aguas_residuales.xlsx")

# Revisar estructura
print(df.info())
display(df.head())

# Convertir fecha
df["fecha_registro"] = pd.to_datetime(df["fecha_registro"])

# Crear indicador de eficiencia de remoción de DBO
df["eficiencia_DBO_%"] = (
    (df["DBO_entrada_mg_L"] - df["DBO_salida_mg_L"]) /
    df["DBO_entrada_mg_L"]
) * 100

# Crear estado textual de cumplimiento
df["estado_cumplimiento"] = df["cumplimiento_norma"].map({
    1: "Cumple",
    0: "No cumple"
})

# Revisar datos finales
display(df.head())
````

````python
# Cumplimiento normativo por planta
cumplimiento = df.groupby("planta")["cumplimiento_norma"].mean().reset_index()

fig1 = px.bar(
    cumplimiento,
    x="planta",
    y="cumplimiento_norma",
    title="Porcentaje de cumplimiento normativo por planta",
    labels={
        "planta": "Planta",
        "cumplimiento_norma": "Cumplimiento promedio"
    }
)

fig1.show()
````

````python
# Evolución temporal de DBO de salida
fig2 = px.line(
    df.sort_values("fecha_registro"),
    x="fecha_registro",
    y="DBO_salida_mg_L",
    color="planta",
    title="Evolución temporal de DBO del efluente tratado",
    labels={
        "fecha_registro": "Fecha",
        "DBO_salida_mg_L": "DBO salida mg/L"
    }
)

fig2.show()
````

````python
# Relación entre caudal y DBO de salida
fig3 = px.scatter(
    df,
    x="caudal_entrada_m3_d",
    y="DBO_salida_mg_L",
    color="planta",
    size="energia_aeracion_kWh",
    hover_data=["fecha_registro", "cumplimiento_norma"],
    title="Relación entre caudal, energía de aireación y DBO de salida"
)

fig3.show()
````

````python
# Distribución de DBO de salida por planta
fig4 = px.box(
    df,
    x="planta",
    y="DBO_salida_mg_L",
    color="planta",
    title="Distribución de DBO de salida por planta"
)

fig4.show()
````

````python
# Eficiencia promedio de remoción de DBO
eficiencia = df.groupby("planta")["eficiencia_DBO_%"].mean().reset_index()

fig5 = px.bar(
    eficiencia,
    x="planta",
    y="eficiencia_DBO_%",
    title="Eficiencia promedio de remoción de DBO por planta"
)

fig5.show()
````

````python
# Archivo para Operaciones
operaciones = df[
    [
        "fecha_registro",
        "planta",
        "caudal_entrada_m3_d",
        "DBO_entrada_mg_L",
        "DBO_salida_mg_L",
        "energia_aeracion_kWh",
        "lodos_generados_kg_d",
        "eficiencia_DBO_%"
    ]
]

operaciones.to_excel("salida_area_operaciones.xlsx", index=False)
files.download("salida_area_operaciones.xlsx")
````

````python
# Archivo para Gestión Ambiental
gestion_ambiental = df[
    [
        "fecha_registro",
        "planta",
        "DBO_salida_mg_L",
        "cumplimiento_norma",
        "estado_cumplimiento"
    ]
]

gestion_ambiental.to_excel("salida_gestion_ambiental.xlsx", index=False)
files.download("salida_gestion_ambiental.xlsx")
````

````python
# indicadores.py

def calcular_eficiencia(df):

    df["eficiencia_DBO_%"] = (
        (df["DBO_entrada_mg_L"] - df["DBO_salida_mg_L"])
        / df["DBO_entrada_mg_L"]
    ) * 100

    return df
````

````python
# visualizaciones.py

import plotly.express as px

def grafico_cumplimiento(df):

    resumen = (
        df.groupby("planta")["cumplimiento_norma"]
        .mean()
        .reset_index()
    )

    fig = px.bar(
        resumen,
        x="planta",
        y="cumplimiento_norma",
        title="Cumplimiento normativo por planta"
    )

    fig.show()
````
